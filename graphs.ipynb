```python
import matplotlib.pyplot as plt
import numpy as np

from mcmc import MetropolisHastingsBracket
from utils import bracket_0, rfc_W
```

```python
ITERS = 20
```

```python
for _ in range(ITERS):
    X = MetropolisHastingsBracket(bracket_0(), win_matrix=rfc_W(), simulate_anneal=True).run(1500)
    plt.semilogy([i.score() for i in X])
plt.ylabel("Likelihood")
ax2 = plt.twinx()
ax2.plot(500000 * .99**np.arange(1500))
ax2.set_ylim(-600, 6000)
ax2.set_ylabel("Temperature")
plt.title("Simulated Annealing Convergence")
plt.xlabel("Iteration")
plt.show()
```
